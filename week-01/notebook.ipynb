### Learning Goal
- **What concept is this notebook teaching?** Python, Numpy, and Pandas basics for Data Science.
- **Why does it matter?** You cannot build ML models if you cannot manipulate data arrays.

```python
import sqlite3
import pandas as pd

df = pd.read_csv('../../data/netflix_titles.csv')
conn = sqlite3.connect(':memory:')
df.to_sql("netflix", conn, index=False)

# df.info()

all_titles = pd.read_sql_query(""" 
    SELECT * FROM netflix
""", conn)

print(all_titles)

stranger_things_details = pd.read_sql_query(""" 
    SELECT * FROM netflix WHERE title="Stranger Things"
""", conn)

print(stranger_things_details)

movies_after_2020 = pd.read_sql_query(""" 
    SELECT title AS Movies_After_2020
    FROM netflix
    WHERE release_year > 2020
      AND type = 'Movie'
    ORDER BY date_added DESC
    LIMIT 100
""", conn)

print(movies_after_2020)

david_fincher_titles = pd.read_sql_query(""" 
    SELECT *
    FROM netflix
    WHERE director LIKE 'David Fincher'
""", conn)

print(david_fincher_titles)

tv_show_length_categories = pd.read_sql_query(""" 
    SELECT COUNT(*) AS Number_of_Shows, 
      CASE 
          WHEN duration = '1 Season' OR duration = '2 Seasons' THEN 'SHORT'
          WHEN duration = '3 Seasons' OR duration = '4 Seasons' THEN 'MODERATE'
          WHEN duration = '5 Seasons' THEN 'LONG'
          ELSE 'VERY HIGH'
      END AS Series_Length
    FROM netflix
    WHERE type = 'TV Show'
    GROUP BY Series_Length;
""", conn)

print(tv_show_length_categories)

long_running_tv_shows = pd.read_sql_query("""
    SELECT title, duration
    FROM netflix
    WHERE type = 'TV Show'
      AND duration IN (
          '8 Seasons',
          '9 Seasons',
          '10 Seasons',
          '11 Seasons',
          '12 Seasons',
          '13 Seasons',
          '14 Seasons',
          '15 Seasons',
          '16 Seasons',
          '17 Seasons',
          '18 Seasons'
      )
""", conn)

print(long_running_tv_shows)

garfield_fincher_movies = pd.read_sql_query("""
    SELECT title, [cast]
    FROM netflix
    WHERE director = 'David Fincher'
      AND [cast] LIKE '%Garfield%';
""", conn)

print(garfield_fincher_movies)
```

### Common Mistakes
- **Beginner Mistake:** Iterating through rows in a Pandas DataFrame using `.iterrows()` instead of using vectorized column operations.

### Practical Takeaway
- **Where does this concept appear in industry?** Every single ML pipeline begins with Pandas/Numpy data wrangling.

### Key Insight
- **Memorable lesson:** Never use a for-loop on a dataframe if a vectorized function exists.
